**6. Display a pivot table reporting average age for each race (by row) and each gender (by column). Then find the maximum age of Amer-Indian-Eskimo men** (2 marks)

```python
# You code here
import pandas as pd
df = pd.read_csv('adults.csv')

pivot = df.pivot_table(values='age', index='race', columns='sex', aggfunc='mean')
max_age = df[(df['race'] == 'Amer-Indian-Eskimo') & (df['sex'] == 'Male')]['age'].max()

print(pivot)
print("Maximum age of Amer-Indian-Eskimo men:", max_age)
```

**7. พิจารณาเฉพาะกลุ่มผู้ชายที่ไม่มีคู่ (สถานะเป็น Never-married, Separated, Divorced, หรือ Widowed) ให้หาสัดส่วนเป็นร้อยละของผู้ที่มีรายได้สูง (>50K) กับผู้ที่มีรายได้ต่ำ** (2 marks)

```python
# You code here
import pandas as pd
df = pd.read_csv('adults.csv')

sadMales = df[(df['sex'] == 'Male') & (df['marital-status'].isin(['Never-married', 'Separated', 'Divorced', 'Widowed']))]
sadMalesTotal = sadMales.shape[0]
sadMalesRich = sadMales[sadMales['salary'] == '>50K'].shape[0]
sadMalesPoor = sadMales[sadMales['salary'] == '<=50K'].shape[0]

print('Total sad males: ', sadMalesTotal)
print('Total rich sad males: ', sadMalesRich, '|', (sadMalesRich/sadMalesTotal*100), '%')
print('Total poor sad males: ', sadMalesPoor, '|', sadMalesPoor/sadMalesTotal*100, '%')
```

**8. หาสัดส่วนเป็นร้อยละของผู้ชายที่มีสถานะเป็น Married-civ-spouse และร้อยละของผู้หญิงที่มีสถานะเป็น Married-civ-spouse status** (2 marks)

**10. ในข้อนี้ให้พิจารณาเฉพาะประเทศ Thailand, Japan, Germany, China, United-States, และ Singapore. ให้หาจำนวนชั่วโมงทำงานเฉลี่ยของคนที่มีรายได้น้อย (ไม่เกิน 50K) กับของที่คนมีรายได้สูง (>50K) ในแต่ละประเทศ โดยรายงานออกมาเป็น 2 คอลัมน์ (5 บรรทัด)** (2 marks)

**11. แสดงจำนวนชั่วโมงทำงานเฉลี่ยของกลุ่มคนโดยแบ่งตามช่วงอายุและเชื้อชาติ (race) ทั้งนี้ช่วงอายุให้แบ่งเป็น [18-30), [30-40), [40-50), [50-60) และ [60-70] เป็นรายบรรทัดและเชื้อชาติเป็นแนวคอลัมน์** (2 marks)

```python
import pandas as pd
df = pd.read_csv('adults.csv')

df['age_group'] = pd.cut(df['age'], bins=[18, 30, 40, 50, 60, 70], right=False, labels=['[18-30)', '[30-40)', '[40-50)', '[50-60)', '[60-70)'])
pivot = df.pivot_table(values='hours-per-week', index='age_group', columns='native-country', aggfunc='mean')

pivot
```

**12. จากข้อที่แล้ว ให้แสดงคอลัมน์ Other ไว้ขวาสุดและค่าเฉลี่ยให้แสดงด้วยจุดทศนิยมสองตำแหน่ง ทำนองดังต่อไปนี้** (1 mark)


```python
# You code here
import pandas as pd
df = pd.read_csv('adults.csv')

df['age_group'] = pd.cut(df['age'], bins=[18, 30, 40, 50, 60, 70], right=False, labels=['[18-30)', '[30-40)', '[40-50)', '[50-60)', '[60-70)'])
pivot = df.pivot_table(values='hours-per-week', index='age_group', columns='native-country', aggfunc='mean')
pivot = pivot.assign(Other = pivot.mean(axis=1))
pivot = pivot.round(2)

pivot
```
