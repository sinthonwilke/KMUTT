#### 1. Gray Level Transformation

```python
# Gray Level Transformation
import numpy as np
import cv2
import matplotlib.pyplot as plt

def GLT(image, transform, coeff=1.0, gamma=1.0):
    table = np.arange(256, dtype=np.uint8)

    if transform == 'negative':
        table = 255 - table

    elif transform == 'identity':
        pass

    elif transform == 'log':
        table = (10 * coeff * np.log(1 + table)).astype(np.uint8)

    elif transform == 'invlog':
        table = (10 * coeff / (np.exp(1 + table) + 1)).astype(np.uint8)

    elif transform == 'root':
        invGamma = 1.0 / gamma
        table = (coeff * ((table / 255.0) ** invGamma) * 255).astype(np.uint8)

    elif transform == 'power':
        table = (coeff * ((table / 255.0) ** gamma) * 255).astype(np.uint8)

    return cv2.LUT(image, table)

original = cv2.imread('original.jpg', )
transform = 'power'
coeff = 1.0
gamma = 2.5
output = GLT(original, transform, coeff, gamma)


original = cv2.imread('original.jpg', cv2.COLOR_BAYER_BG2GRAY)
transform = 'power'
coeff = 1.0
gamma = 3.0
output = GLT(original, transform, coeff, gamma)


plt.figure(figsize=(10, 10))

plt.subplot(121)
plt.imshow(original)
plt.title('Original')
plt.axis('off')

plt.subplot(122)
plt.imshow(output)
plt.title('Output')
plt.axis('off')

plt.show()
```

#### 2. Piecewise-Linear Transformations

```python
# Piecewise-Linear Transformations
import numpy as np
import cv2

def piecewise_linear_transform(image, breakpoints, slopes):
    lut = np.arange(256, dtype=np.uint8)

    for i in range(len(breakpoints) - 1):
        start_point, end_point = breakpoints[i], breakpoints[i + 1]
        slope = slopes[i]

        for x in range(start_point, end_point):
            lut[x] = int(slope * (x - start_point)) 
            
    return cv2.LUT(image, lut)


original = cv2.imread('original.jpg', cv2.COLOR_BAYER_BG2GRAY)
breakpoints = [0, 128, 230]
slopes = [1, 2, 1]
output = piecewise_linear_transform(original, breakpoints, slopes)


plt.figure(figsize=(10, 10))

plt.subplot(121)
plt.imshow(original)
plt.title('Original')
plt.axis('off')

plt.subplot(122)
plt.imshow(output)
plt.title('Output')
plt.axis('off')

plt.show()
```

#### 3. Gray Level Slicing (choose your interest part to slice)

```python
# Gray Level Slicing (choose your interest part to slice)
import numpy as np
import cv2

def GLS(image, lower, upper):
    lut = np.arange(256, dtype=np.uint8)
    lut[lower:upper + 1] = 255
    return cv2.LUT(image, lut)

original = cv2.imread('original.jpg', cv2.COLOR_BAYER_BG2GRAY)
lower = 235
upper = 255
output = GLS(original, lower, upper)

plt.figure(figsize=(10, 10))

plt.subplot(121)
plt.imshow(original)
plt.title('Original')
plt.axis('off')

plt.subplot(122)
plt.imshow(output)
plt.title('Output')
plt.axis('off')

plt.show()
```

#### 4. Bit Plane Slicing

```python
# Bit Plane SlicingÃŸ
import cv2
import numpy as np
import matplotlib.pyplot as plt

def extract_bit_plane(image, bit_position):
    bit_plane = (image >> bit_position) & 1
    bit_plane = bit_plane * 255
    return bit_plane

original = cv2.imread('original.jpg', cv2.COLOR_BAYER_BG2GRAY)

fig, axs = plt.subplots(1, 9, figsize=(20, 5))
axs[0].imshow(original)
axs[0].set_title('Original')
axs[0].axis('off')
for i in range(8):
    bit_plane = extract_bit_plane(original, i)
    axs[i+1].imshow(bit_plane)
    axs[i+1].set_title('Bit Plane {}'.format(i))
    axs[i+1].axis('off')
plt.show()
```

#### 5. Histogram Equalization

```python
# Histogram Equalization
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('original.jpg', cv2.IMREAD_GRAYSCALE)
equalized_image = cv2.equalizeHist(image)

plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.imshow(image, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(122)
plt.imshow(equalized_image, cmap='gray')
plt.title('Output')
plt.axis('off')

plt.show()
```

#### 6. Smooth filter/sharpen filter

##### Smooth filter - Gaussian

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('original.jpg', cv2.COLOR_BAYER_BG2GRAY)

kernel_size = (5, 5)
blurred_image = cv2.GaussianBlur(image, kernel_size, 0)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.title('Blurred Image (Gaussian Blur)')
plt.axis('off')

plt.show()

```

##### Shapen filter - Laplacian

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('original.jpg', cv2.COLOR_BAYER_BG2GRAY)

sharpened_image = cv2.Laplacian(image, cv2.CV_64F)
sharpened_image = cv2.convertScaleAbs(sharpened_image)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sharpened_image, cmap='gray')
plt.title('Sharpened Image (Laplacian)')
plt.axis('off')

plt.show()
```
